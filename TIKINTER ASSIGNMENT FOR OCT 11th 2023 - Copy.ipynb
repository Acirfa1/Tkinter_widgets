```python
# setting the font style using the font object of tkinter.font
# BREAKING DOWN THE APPROACH
# Import the tkinter module
# Import tkinter font
# Create the GUI window
# Create our text widget
# Create an object of type Font from tkinter.font module. It takes in the desired font specifications
# (font_family, font_size_in_pixel, font_weight) as a constructor of this object. This is that 
# specified object that the text widget requires while determining its font.
# Parse the Font object to the Text widget using configure() method.



# BELOW IS THE IMPLEMENTATION OF THE ABOVE APPROACH:







# Using Tkinter to change the font style
import tkinter as tk
from tkinter import *
import tkinter.font as tkFont


#main window
root = tk.Tk()
#title of the window
root.title("This is a program for change in font size and I hope it works")

#creating our text widget.
sample_text=tk.Text(root, height = 10)
sample_text.pack()


#creating an object of type Font from tkinter. 
Desired_font = tkFont.Font( family = "Times", size = 10, weight = "bold")

# Parsed the font object
# to the text widget using .configure() method.
sample_text.configure(font = Desired_font)

root.mainloop()
```

```python
# The padding options padx and pady of the grid and pack methods can take a 2-tuple that represent 
# the left/right and top/bottom padding




# Import required libraries
from tkinter import *

# Create an instamnce of tkinter window
parent = Tk()






#parent.geometry("700x350")

# Degfine the padding valuesin variables
px = 30
py = 20

# Create a frame widget
frame =Frame(parent, width=40, height=65, bg= "green")


# Create a label widget
label = Label(frame, text="This is an example of Padding", font=("Arial 15 bold"), bg= "skyblue")
label.pack(padx=px, pady=py)
frame.pack(padx=px, pady=py)

parent .mainloop()
```

```python
# To include radio button and checkbox in tkinter
# Radio button in tkinter



import tkinter as tk

window = tk.Tk()
window.title("My Window")
window.geometry("500x300")

var= tk.StringVar()
label = tk.Label(window,font= "Arial", bg="blue", width=40, text= "This is me working on a tkinter radio button")
label.pack()


def sel():
    selection = "You selected the correct option " + str(var.get())
    label.config(text = selection)
r1= tk.Radiobutton(window, text="option 1", variable=var, value=1, command=sel)
r1.pack( anchor = W )
r2= tk.Radiobutton(window, text="option 2", variable=var, value=2, command=sel)
r2.pack( anchor = W )
r3= tk.Radiobutton(window, text="option 3", variable=var, value=3, command=sel)
r3.pack( anchor = W )
label = Label(window)
label.pack()

window.mainloop()
```

```python
import tkinter as tk

check_box = tk.Tk()
check_box.title("This is me trying to work on a switch for checkbox in a widget")
check_box.geometry("100x100")

label = tk.Label(check_box, bg = "blue", width = 20, text = "empty")

def print_selection():
    if (var1.get() == 1) and (var2.get() == 0):
        label.config(text="I love Data Science  ")
    elif (var1.get() == 0) and (var2.get() == 1):
        label.config(text="I love Barcelona ")
    elif (var1.get() == 0) and (var2.get() == 0):
        label.config(text="I hate Chelsea and Manu and Madrid too")
    else:
        label.config(text="I love Data Science and Barcelona")
        
        
var1 = tk.IntVar()
var2 = tk.IntVar()
c1 = tk.Checkbutton(check_box, text="Data Science", variable=var1, onvalue=1, offvalue=0, command=print_selection)
c1.pack()
c2 = tk.Checkbutton(check_box, text="Barcelona", variable=var2, onvalue=1, offvalue=0, command=print_selection)
c2.pack()

check_box.mainloop()
```
